```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 12,9
```

```python
sales = pd.read_csv('abd_4.csv')
sales.head()
```

```python
films_ = sales.filter(items = ['Название фильма', 'Сумма'])
films = films_.groupby(['Название фильма']).sum()
films.head()
```

```python
films['Сумма'].plot(kind='pie', autopct='%1.1f%%', 
                                 startangle=90, shadow=False, legend = True, fontsize=14)
```

```python
#магазин 1, top5
films_magazin1 = sales[sales['Магазин'] == "Магазин 1"].sort_values('Сумма', ascending=False)
films_from_magazin1_ = films_magazin1.filter(items = ['Название фильма', 'Сумма'])
films_from_magazin1 = films_from_magazin1_.groupby(['Название фильма']).sum()
films_from_magazin1.sort_values(['Сумма'], ascending=False).head(5)
```
