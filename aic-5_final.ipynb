```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind as ttest
```

```python
df = pd.read_csv('datasale.csv')
df.head()
```

```python
df.info()
```

приведем в порядок датафрейм

```python
df = df.dropna()
```

```python
df['Quantity']=df['Quantity'].astype(int)
df['UnitPrice'] = [x.replace(',', '.') for x in df['UnitPrice']] 
df['UnitPrice'] = df['UnitPrice'].astype(float) 
```

Проведем анализ по странам и посмотрим по продукции что пользуется популярностью, а что - нет.

выручка и количество проданных единиц по странам, топ-10 и антитоп-10

```python
df_country = df.groupby(['Country']).sum().sort_values(by='UnitPrice', ascending=False).head(10)
df_country['Country'] = df_country.index
df_country.to_csv('df_country.csv', index=False)
df_country
```

Что и стоило ожидать - по ЕС работа ведется бойко

```python
df_anticountry = df.groupby(['Country']).sum().sort_values(by='UnitPrice', ascending=True).head(10)
df_anticountry['Country'] = df_anticountry.index
df_anticountry.to_csv('df_anticountry.csv', index=False)
df_anticountry
```

Если по таблице выше основные покупатели - ЕС, то почему-то проблемы есть у Чехии и Латвии. С остальными понятно - далеко, доставка дорогая, ожидание долгое.

ТОП-10 и антитоп-10 самых продаваемых вещей

```python
df_sale = df.groupby(['Description', 'StockCode']).sum().sort_values(by='Quantity', ascending=False).head(10)
df_sale['StockCode'] = df_sale.index
df_sale.to_csv('df_sale.csv', index=False)
df_sale
```

Похоже, что JUMBO BAG RED RETROSPOT нынче в моде))

```python
df_antisale = df.groupby(['Description', 'StockCode']).sum().sort_values(by='Quantity', ascending=True).head(10)
df_antisale['StockCode'] = df_antisale.index
df_antisale.to_csv('df_antisale.csv', index=False)
df_antisale
```

Большие проблемы с заказами по продукции выше. Если это брошенные корзины, то, возможно, наша цена выше среднего. Если возвраты - то, вероятно, с продукцией в серии что-то не то.
